```python
from scipy import stats
import numpy as np

# Calculate Z-scores
z_scores = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))

# Set a threshold for outliers (commonly 3)
outliers = (z_scores > 3).sum(axis=1)

# Show rows with outliers
outliers_rows = df[outliers > 0]
print(outliers_rows)
```

```python
average_price = df['price'].mean()
print(f"Average Price of Games: ${average_price:.2f}")
```

```python
average_metacritic_score = df['metacritic_score'].mean()
print(f"Average Metacritic Score: {average_metacritic_score:.2f}")

```

```python
top_10_metacritic = df[['name', 'metacritic_score']].sort_values(by='metacritic_score', ascending=False).head(10)
print("Top 10 Games by Metacritic Score:")
print(top_10_metacritic)
```

```python
top_10_reviews = df[['name', 'num_reviews_total']].sort_values(by='num_reviews_total', ascending=False).head(10)
print("Top 10 Games by Number of Reviews:")
print(top_10_reviews)
```

```python
top_10_expensive = df[['name', 'price']].sort_values(by='price', ascending=False).head(10)
print("Top 10 Most Expensive Games:")
print(top_10_expensive)
```

```python
# a composite score based on Metacritic score and number of reviews, we are trying to identify the top 10 games that are both highly rated and widely popular
# Normalize scores (e.g., scale to [0, 1])
df['metacritic_norm'] = (df['metacritic_score'] - df['metacritic_score'].min()) / (df['metacritic_score'].max() - df['metacritic_score'].min())
df['reviews_norm'] = (df['num_reviews_total'] - df['num_reviews_total'].min()) / (df['num_reviews_total'].max() - df['num_reviews_total'].min())

# Create a weighted composite score (you can adjust the weights)
df['composite_score'] = 0.6 * df['metacritic_norm'] + 0.4 * df['reviews_norm']

# Top 10 games by composite score
top_10_composite = df[['name', 'composite_score']].sort_values(by='composite_score', ascending=False).head(10)
print("Top 10 Games by Composite Score (Metacritic & Reviews):")
print(top_10_composite)
```

```python
# Distribution of games targeted at different age groups.)
age_categories = {
    'Kids': (df['required_age'] <= 12),
    'Teens': (df['required_age'] > 12) & (df['required_age'] <= 17),
    'Adults': (df['required_age'] >= 18)
}

# Count the number of games in each category
games_by_age = {category: len(df[condition]) for category, condition in age_categories.items()}
print(games_by_age)
```

```python
from datetime import datetime

# Convert the release_date column to datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Calculate the age of each game (in years)
current_year = datetime.now().year
df['game_age'] = current_year - df['release_date'].dt.year

# Calculate the average age of games
average_game_age = df['game_age'].mean()
print(f"Average Age of Games: {average_game_age:.2f} years")

```

```python
import matplotlib.pyplot as plt

# Plot a histogram of game ages
plt.hist(df['game_age'].dropna(), bins=20, edgecolor='black')
plt.title('Distribution of Game Ages')
plt.xlabel('Game Age (Years)')
plt.ylabel('Number of Games')
plt.show()
```

```python

import seaborn as sns

# Select numeric columns for correlation analysis
numeric_columns = ['price', 'num_reviews_total', 'metacritic_score', 'game_age', 'pct_pos_recent']
corr_matrix = df[numeric_columns].corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()


```
