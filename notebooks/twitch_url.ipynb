```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataframe df is already created and has the necessary columns
# Grouping data by 'Year' and 'Month', and summing 'Hours_watched' for each group
monthly_data = df.groupby(['Year', 'Month'])['Hours_watched'].sum().reset_index()

# Filter the data for years 2016 to 2024
monthly_data = monthly_data[(monthly_data['Year'] >= 2016) & (monthly_data['Year'] <= 2024)]

# Set the figure size with a black background
plt.figure(figsize=(12, 7), facecolor='black')

# Plotting the data with pink color for the line
plt.plot(monthly_data['Year'].astype(str) + '-' + monthly_data['Month'].astype(str), 
         monthly_data['Hours_watched'], 
         marker='o', 
         color='#FF69B4',  # Pink color
         label='Hours Watched', linewidth=2)

# Customize the axes with white text for readability
plt.xlabel('Year-Month', fontsize=14, color='white')
plt.ylabel('Total Hours Watched', fontsize=14, color='white')

# Title with pink color
plt.title('Total Hours Watched by Month (2016-2024)', fontsize=16, color='#FF69B4')  # Pink
plt.xticks(rotation=45, fontsize=12, color='white', ha='right')  # Rotate labels and align them
plt.yticks(fontsize=12, color='white')

# Customize gridlines with a subtle touch
plt.grid(True, linestyle='--', color='white', alpha=0.5)

# Display the plot with tight layout to adjust the space
plt.tight_layout()

# Show the plot
plt.show()
```
